```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ãƒ‡ãƒ¼ã‚¿ã®èª­ã¿è¾¼ã¿
df = pd.read_csv("C:\\Users\\soshi\\Documents\\MPC\\simulation_results.csv")
df=df[:226]
# ãƒ©ã‚¸ã‚¢ãƒ³ã®ã¾ã¾å‡¦ç†
angle1 = df['theta1']  # è§’åº¦ï¼ˆãƒ©ã‚¸ã‚¢ãƒ³ï¼‰
angular_velocity1 = df['theta1_dot']  # è§’é€Ÿåº¦ï¼ˆãƒ©ã‚¸ã‚¢ãƒ³/ç§’ï¼‰

# å¹³å‡ã‚’å¼•ã„ã¦ä¸­å¿ƒã‚’åŽŸç‚¹ã«æƒãˆã‚‹
angle1_centered = angle1 - np.mean(angle1)
angular_velocity1_centered = angular_velocity1 - np.mean(angular_velocity1)

# ã‚°ãƒ©ãƒ•ã®ä½œæˆ
plt.figure(figsize=(6, 6))
plt.plot(angle1_centered, angular_velocity1_centered, 'k-', label="Simulated Trajectory")  # é»’å®Ÿç·š


# è»¸ãƒ©ãƒ™ãƒ«ã¨ç¯„å›²è¨­å®š
plt.xlabel("Position (rad)")
plt.ylabel("Velocity (rad/s)")
plt.title("Phase Plane: Theta1 vs Theta1_dot")

# è»¸ã‚’ä¸­å¿ƒã«è¨­å®š
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # yè»¸ã®ä¸­å¿ƒç·š
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # xè»¸ã®ä¸­å¿ƒç·š

# å‡¡ä¾‹ã¨ã‚°ãƒªãƒƒãƒ‰
plt.legend()
plt.grid()

# ã‚°ãƒ©ãƒ•ã®è¡¨ç¤º
plt.show()
```
