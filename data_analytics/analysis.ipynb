```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv("0.0873_3060_without_noise_delay.csv")
df2 = pd.read_csv("0.2618_3060_without_noise_delay.csv")
```

# 足首トルクの位相プロット

```python

# プロットの作成
plt.figure(figsize=(8, 6))

# df1のプロット（黒色）
plt.plot(df1['q1'][:4970], df1['dq1'][:4970], linestyle='-', color='black', label='df1')

# df2のプロット（青色）
plt.plot(df2['q1'][:4970], df2['dq1'][:4970], linestyle='-', color='red', label='df2')

# 目標点 (0,0) を赤い点で表示
plt.plot(0, 0, 'bo')  # 'r' は赤色、'o' は丸（ドット）

# 軸ラベルとタイトル
plt.xlabel('Position (rad)')
plt.ylabel('Velocity (rad/s)')
plt.title('MPC')

# 凡例の表示
plt.legend()

# グリッド表示
plt.grid(True)

# プロットの表示
plt.show()
```

```python
# q1 vs dq1 の2次元プロット（範囲指定あり）
plt.figure(figsize=(8, 6))
plt.plot(df1['q1'][:4970], df1['dq1'][:4970], linestyle='-',color='black')
plt.plot(0, 0, 'bo')  # 'r' は赤色、'o' は丸（ドット）

plt.xlabel('Position(rad)')
plt.ylabel('Velocity(rad/s)')
plt.title('MPC')

plt.legend()
plt.grid(True)
plt.show()
```

```python
# q1 vs dq1 の2次元プロット（範囲指定あり）
plt.figure(figsize=(8, 6))
plt.plot(df2['q1'][:4970], df2['dq1'][:4970], linestyle='-',color='red', label='q1 vs dq1')
plt.plot(0, 0, 'bo')  # 'r' は赤色、'o' は丸（ドット）

plt.xlabel('Position(rad)')
plt.ylabel('Velocity(rad/s)')
plt.title('MPC')
plt.legend()
plt.grid(True)
plt.show()
```
